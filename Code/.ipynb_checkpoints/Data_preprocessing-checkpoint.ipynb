```python
#### INPUTEVENTS_CV
INPUTEVENTS_CV=pd.read_csv(dpath+'INPUTEVENTS_CV.csv')
inp_cv=INPUTEVENTS_CV['subject_id'].isin(pateinetid.values)
df_inpcv=INPUTEVENTS_CV[inp_cv]
df_inpcv=df_inpcv.drop_duplicates(subset = ["subject_id"])
# df_inpcv.head(3)
dfinpucv=df_inpcv[['subject_id','amount']]
print(df_inpcv.shape)
dfinpucv.head(3)
```

```python
### INPUTEVENTS_MV
INPUTEVENTS_MV=pd.read_csv(dpath+'INPUTEVENTS_MV.csv')
inp_mv=INPUTEVENTS_MV['subject_id'].isin(pateinetid.values)
df_inpmv=INPUTEVENTS_MV[inp_mv]
df_inpmv=df_inpmv.drop_duplicates(subset = ["subject_id"])
df_inpmv.head(3)
dfinpumv=df_inpmv[['subject_id','amount']]
print(df_inpmv.shape)
dfinpumv.head(3)
```

```python
### MICROBIOLOGYEVENTS
MICROBIOLOGYEVENTS=pd.read_csv(dpath+'MICROBIOLOGYEVENTS.csv')
mic=MICROBIOLOGYEVENTS['subject_id'].isin(pateinetid.values)
df_mic=MICROBIOLOGYEVENTS[mic]
df_mic=df_mic.drop_duplicates(subset = ["subject_id"])
df_mic.head(3)
dfmic=df_mic[['subject_id','spec_type_desc']]
print(dfmic.shape)
dfmic.head(3)
dfmic['spec_type_desc'].value_counts()
```

```python
### PROCEDUREEVENTS_MV
PROCEDUREEVENTS_MV=pd.read_csv(dpath+'PROCEDUREEVENTS_MV.csv')
pro=PROCEDUREEVENTS_MV['subject_id'].isin(pateinetid.values)
df_pro=PROCEDUREEVENTS_MV[pro]
df_pro=df_pro.drop_duplicates(subset = ["subject_id"])
df_pro.head(3)
dfpro=df_pro[['subject_id','value']]
print(dfpro.shape)
dfpro.head(3)
# dfpro['value'].value_counts()
```

```python
### SERVICES
SERVICES=pd.read_csv(dpath+'SERVICES.csv')
ser=SERVICES['subject_id'].isin(pateinetid.values)
df_ser=SERVICES[ser]
df_ser=df_ser.drop_duplicates(subset = ["subject_id"])
df_ser.head(3)
dfser=df_ser[['subject_id','curr_service']]
print(dfser.shape)
dfser.head(3)
# dfser['curr_service'].value_counts()
```

```python
data_frames=[dfeth,dfchar,dfinpucv, dfinpumv,dfmic, dfpro,dfser,y]
from functools import reduce
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['subject_id'],
                                            how='outer'), data_frames)
print(df_merged.shape)
df_merged.head(3)
```
